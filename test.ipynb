```python
import numpy as np
import pandas as pd
import os
import glob 
label_columns=["sofifa_id","ls","st","rs","lw","lf","cf","rf","rw","lam","cam","ram","lm","lcm","cm","rcm","rm","lwb","ldm","cdm","rdm","rwb","lb","lcb","cb","rcb","rb"]

path = 'C:/Users/Gaurav Gandhi/Documents/pattern/fifa20/players_20.csv'
df=pd.read_csv(path,sep=",",header="infer")
label_train=df.filter(items=label_columns)
label_train.update(label_train.fillna("0+0"))
label_train=label_train[label_train.ls.notnull()]

# GOAL KEEPER
def split_1(v):
    if '+' in v:
        s= v.split('+')
        return (int(s[0])+int(s[1]))
    else:
        s=v.split('-')
        return int(s[0])-int(s[1])
df["gk_positioning"]=df["gk_positioning"].apply(lambda v: 0 if type(v)==str and len(v)>4 else( split_1(v) if type(v)== str and len(v)==4 else v))


ids=label_train.sofifa_id
label_train.drop(columns=["sofifa_id"],inplace=True)


for col in label_train.columns[1:]:
    label_train[col]=label_train[col].apply(lambda v:v.split('+')).apply(lambda s: int(s[0])+int(s[1]))
label_train=label_train[label_train.apply(lambda x: x == label_train.max(axis = 1))]

label_train.insert(0,"gk_positioning",df["gk_positioning"],allow_duplicates=True)
label_train.replace(0.0,np.nan,inplace=True)

label_train
```
